```python
#Para filtrar um valor distinto em uma coluna, você pode usar uma máscara booleana no Pandas.
#Por exemplo, para filtrar todas as linhas onde a coluna time é igual a "Dinner", você faria assim:
# Filtrar onde o valor de 'time' é 'Dinner'
filtro = dados['time'] == 'Lunch'
dados_filtrados = dados[filtro]

# Exibindo o DataFrame filtrado
print(dados_filtrados)
```

```python
# Filtrando para obter apenas as linhas onde o valor da coluna 'time' é 'Lunch'
filtro = dados['time'] == 'Lunch'
dados_lunch = dados[filtro]

# Obtendo os índices das linhas filtradas
indices_lunch = dados.index[dados['time'] == 'Lunch'].tolist()

# Exibindo o DataFrame filtrado e os índices
print(dados_lunch)
print("Índices dos almoços:", indices_lunch)
```

```python
#Para filtrar o DataFrame e obter todas as linhas onde sex é "Male" e time é "Lunch", você pode usar o seguinte código:
# Filtrando os dados para obter apenas os 'Males' que tiveram 'Lunch'
filtro_male_lunch = (dados['sex'] == 'Male') & (dados['time'] == 'Lunch')
dados_male_lunch = dados[filtro_male_lunch]

# Exibindo o DataFrame filtrado
print(dados_male_lunch)
```
